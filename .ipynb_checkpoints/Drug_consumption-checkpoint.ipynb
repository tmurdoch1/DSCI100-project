## Preliminary exploratory data analysis

```r
#load library
library(tidyverse)
library(tidymodels)
library(repr)
#set seed
set.seed(1)
```

```r
#read data set
drug_data <- read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00373/drug_consumption.data",col_names = FALSE)
#select needed predictors and class
drug_data_selected <- drug_data |>
     select(X1:X6, X21)
```

```r
#make column names human readable
drug_data_clean <- drug_data_selected |> 
       rename(identifier = X1, age = X2, gender = X3,
       education = X4, country = X5, ethnicity = X6,  cocaine_usage = X21)

#change column "use" into factor
drug_data_clean <- drug_data_clean |>       
       mutate(cocaine_usage = as_factor(cocaine_usage))
```

Now the data is clean and in tidy format, we can print the data.

```r
#print data
print(drug_data_clean)
```

We split data into training and testing data with 75% being training data.

```r
#split data
drug_split <- initial_split(drug_data_clean, prop = 0.75, strata = cocaine_usage)
drug_train <- training(drug_split)
drug_test <- testing(drug_split)
```

### Summarize the data
To give an overview of the training data:
1. summarize the mean value for each predictors
2. count the number of observation for each class
3. find out number of missing row

```r
#summarize mean and number of observation in a table
data_summarize <- drug_train |>
    group_by(cocaine_usage) |>
    summarize(age_mean = mean(age), gender_mean = mean(gender), education_mean = mean(education),
              country_mean = mean(country), ethnicity_mean = mean(ethnicity), observations = n())
```

Now we summarized data within one table, where first column reprecents class, colums 2-5 indicates the mean value of each predictors, last column show the number of observations of each class.

```r
data_summarize
```

```r
#find out number of missing row

rowSums(is.na(drug_train))
```

### Visualize the data
Since our data is standardized, the scatter plot and regular bar plot doesn't presents vairable distribution very well. Instead we use function "fill" to eualize the height of bar plot so we can compare the proportion of class based on different predictors, which allow us to find which varaibles are useful to distinguish wether a person being a drug user.

```r
#Plot to see trend of use in variables
age_plot <- drug_train |>
    ggplot(aes(x = age, fill = cocaine_usage)) + 
    geom_bar(width = 0.5, position = "fill") +
    xlab("Age from low to high, 18 to 65+ (standardized)") +
    ylab("Proportion") +
    theme(text = element_text(size = 15))

gender_plot <- drug_train |>
    ggplot(aes(x = gender, fill = cocaine_usage)) + 
    geom_bar(position = "fill") +
    xlab("Gender: Male(left), Female(right)") +
    ylab("Proportion") +
    theme(text = element_text(size = 15))

education_plot <- drug_train |>
    ggplot(aes(x = education, fill = cocaine_usage)) + 
    geom_bar(width = 0.2, position = "fill") +
    xlab("Education achievement low to high (standardized)") +
    ylab("Proportion") +
    theme(text = element_text(size = 15))

country_plot <- drug_train |>
    ggplot(aes(x = country, fill = cocaine_usage)) + 
    geom_bar(width = 0.03, position = "fill") +
    xlab("Country (standardized)") +
    ylab("Proportion") +
    theme(text = element_text(size = 15))

ethnicity_plot <- drug_train |>
    ggplot(aes(x = ethnicity, fill = cocaine_usage)) + 
    geom_bar(position = "fill") +
    xlab("Ethnicity (standardized)") +
    ylab("Proportion") +
    theme(text = element_text(size = 15))
```

```r
age_plot
options(repr.plot.width = 10, repr.plot.height = 4)
```

As plot shows, age is a strong indicator of being a cocaine user. The likelihood of being a cocaine user decrease as age increase.

```r
gender_plot
options(repr.plot.width = 6, repr.plot.height = 4)
```

Gender is a strong indicator since being a male increase the chance of being a cocaine user.

```r
education_plot
options(repr.plot.width = 10, repr.plot.height = 4)
```

As plot shows, the education achievement tends to decrease drug consumption. Which the higher the education achievements, more likely 'never used' cocaine.

```r
country_plot
options(repr.plot.width = 10, repr.plot.height = 5)
```

As plot shows, country affects wether a person being a cocaine user. 

```r
ethnicity_plot
```

Ethinicity give strong indication of cocaine usage, which being an African-American are morelike to use cocaine.

In conclusion of above discussion, we decide to use `age`, `gender`, `education`, `country` and `ethnicity` as predictors.
